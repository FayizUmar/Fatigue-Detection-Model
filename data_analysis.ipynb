```python
# Load the dataset from the CSV file
data = pd.read_csv("C:/Users/joeja/Downloads/originalds.csv")

# Extract the age column
age_data = data['hear_rate']

# Create a histogram
plt.hist(age_data, bins=10, edgecolor='black')

# Set labels and title
plt.xlabel('Heart Rate')
plt.ylabel('Frequency')
plt.title('Frequency Distribution of Heart Rate')

# Display the graph
plt.show()

# Find the minimum age
min_age = data['hear_rate'].min()
print(min_age)

# Find the maximum age
max_age = data['hear_rate'].max()
print(max_age)

avg_age=data['hear_rate'].mean()
print(avg_age)

```
